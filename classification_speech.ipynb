<p style="font-family: Arial; font-size:1.75em;color:purple; font-style:bold"><br>

Holdout - diviser le dataset en ensemble d'apprentissage et test

<br><br></p>



```python
y=df_global[['word']].copy()
```

```python
X=df_global.copy()
del X['word']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

#  ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
extra_trees =ExtraTreesClassifier()
extra_trees.fit(X_train, y_train)
prediction = extra_trees.predict(X_test)
```

```python
acc=accuracy_score(y_true = y_test, y_pred = prediction)
precision=precision_score(y_test,prediction,average=None)
recall=recall_score(y_test,prediction,average=None)
mesure=f1_score(y_test,prediction,average=None)
```

```python
print(acc)
```

# AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
ada_reg = AdaBoostClassifier()
ada_reg.fit(X_train, y_train)
```

```python
prediction_ = ada_reg.predict(X_test)
```

```python
acc_=accuracy_score(y_true = y_test, y_pred = prediction_)
precision_=precision_score(y_test,prediction_,average=None)
recall_=recall_score(y_test,prediction_,average=None)
mesure_=f1_score(y_test,prediction_,average=None)
```

```python
print(acc_)
```
